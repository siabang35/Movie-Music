```python
# modules
import pandas as pd
import pickle 

#importing preprocessed file
movies=pickle.load(open('./data/movies_meta_data.pkl','rb'))

movies
```

```python
# list of id with its title 
id_title=[]

# list of user id
user_id=[]

# iterate over all movies in dataframe
for idx in movies.index:
    id_title.append([ movies['id'][idx],movies['title'][idx] ])
    user_id.append([movies['id'][idx],movies['userId'][idx],movies['rating'][idx]])

# generate dataframe of movie id and title 
id_title_df=pd.DataFrame(id_title,columns=['id','title']).drop_duplicates()

# generate dataframe of movie id , user id and rating 
ueser_id_df=pd.DataFrame(user_id,columns=['id','userId','rating']).drop_duplicates()
```

```python
# generate matrix of user id and movie id with value as rating and repalce all nan value with 0
user_movie_mat=pd.pivot(ueser_id_df,index='userId',columns='id',values='rating').fillna(0)
```

```python


# reset index of user_movie_mat
user_movie_mat=user_movie_mat.reset_index()
```

```python
# generate matrix of user id and movie id with value as rating  again with new index
user_movie_mat=pd.pivot(ueser_id_df,index='id',columns='userID',values='rating').fillna(0)
user_movie_mat
```

```python
# normalise all values (i.e mean centric all values column wise)
def standardize(col):
    new_col=(col-col.mean())/(col.max()-col.min())
    return new_col
rating_std=user_movie_mat.apply(standardize)

# transpose matrix
rating_std.T
```

```python
# import cosine_similarity for calculating angle between vector in multi-dimensional space
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# calculate similarity matrix of each movies with respect to each other movies
similarity = cosine_similarity(user_movie_mat.T)
similarity
```

```python
# convert similarity matrix into data frame
similarity_df=pd.DataFrame(similarity,index=user_movie_mat.T.index,columns=user_movie_mat.T.index)
similarity_df
```

```python
import pickle

# create a file in the directory
pickle.dump(similarity_df,open('item_item_collaborative_sim_mat.pkl','wb'))
```

```python
# it accepts movie name and rating given by user to that movie 
def get_similar_movies(movie_name,user_rarting):
    index=0
    # finding id of movie by its title
    for idx in ueser_id_df.index: 
        if ueser_id_df['title'][idx]==movie_name:
            index=ueser_id_df['id'][idx]
            break
    # calcualting similarity score of each movie again accorind to user rating ( if rating is less than 5 then decrease its similarity score or vice-versa)
    similar_score=similarity_df[index]*(user_rarting-5)

    # sort them all in decreasing order
    similar_score=similar_score.sort_values(ascending=False)

    return similar_score
```

```python
# list of movies and its rating rated by user in real time 
list_of_user_preferences=[["Iron Man",1],["Spider-Man",9],["Spider-Man 3",8]]
similar_movies=pd.DataFrame()

# for each movie recalculate movies similarity according to user preferences
for elm in list_of_user_preferences:
    similar_movies=similar_movies.append(get_similar_movies(elm[0],elm[1]),ignore_index=True)


# suming up all values for each movies and sort them in decreasing order
similar_movies=similar_movies.sum().sort_values(ascending=False)
similar_movies=list(similar_movies.index)
```
