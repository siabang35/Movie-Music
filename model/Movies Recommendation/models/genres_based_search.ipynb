```python
# merge average rating with dataframe of genres and movie id
movieId_geners_avgRating=gener_movie_id.merge(movies_usersRating_df,on='movie_id')
movieId_geners_avgRating
```

```python
# remove rows having null values and remove duplicates rows
movieId_geners_avgRating.dropna(inplace=True)
movieId_geners_avgRating=movieId_geners_avgRating.drop_duplicates()
```

```python
# generate matrix between movies id and genres with values as average rating
movieId_geners_avgRating_mat=movieId_geners_avgRating.pivot(index='movie_id',columns='genres',values='avg_rating').fillna(0)
movieId_geners_avgRating_mat
```

```python
#  this function accepts genres name and number of recommend of movies user want and return a list of id by sorting all movie of same genres with reverse order of average ratings
def recommend_movies_by_genres(genres_name,no_of_recommendations):
    movies_list=movieId_geners_avgRating_mat[genres_name]
    movies_list=movies_list.sort_values(ascending=False)
    recommend_movies=[]
    counter=0
    for i in movies_list.index:
        if(counter==no_of_recommendations+1):
            break
        recommend_movies.append(i)
        counter=counter+1
    return recommend_movies    
recommend_movies_by_genres('Adventure',5)
```
