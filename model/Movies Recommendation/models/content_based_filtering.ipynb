```python
# Modules
import pandas as pd
import ast
import pickle 

#importing preprocessed file
movies=pickle.load(open('./data/all_movies.pkl','rb'))

# remove spaces from string
def removeSpaces(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
movies
```

```python

# for finding belongs to in colletions
def find_name_in_belongstocollection(iter):
    return (ast.literal_eval(iter)['name'])



movies['belongs_to_collection']=movies['belongs_to_collection'].apply(find_name_in_belongstocollection)
movies['belongs_to_collection']=movies['belongs_to_collection'].apply(lambda x:x.split(" "))

# convert list of collection name to a stingle string
for idx in movies['belongs_to_collection'].index:
    str=""
    for word in movies['belongs_to_collection'][idx]:
        str+=word
    movies['belongs_to_collection'][idx]=str

movies['belongs_to_collection']
```

```python
# convert string to list of string
movies['title']=movies['title'].apply(lambda x:x.split())
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies['tagline']=movies['tagline'].apply(lambda x:x.split())
movies['belongs_to_collection']=movies['belongs_to_collection'].apply(lambda x:x.split())

movies
```

```python
# merge all string of all columns into overview of each movies
movies['overview']=movies['genres'] + movies['title'] + movies['overview'] +movies['cast']+movies['crew']+movies['keywords']+movies['tagline']
movies['overview']=movies['overview']+movies['belongs_to_collection']
movies
```

```python
# filter out id and overview  into new dataframe
new=movies[['id','overview']]
new
```

```python
# convert string to list of string
new['overview']=new['overview'].apply(lambda x: " ".join(x))

# remove duplicate rows
new= new.drop_duplicates()

new
```

```python
# this is for fuzzy matching
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new['overview']=new['overview'].apply(stem)
```

```python
# create a file in directory
movies_with_title=pickle.load(open('./data/all_movies.pkl','rb'))
movies_with_title
```

```python
# list of movie names
movies_list=[]

# list of id with title of movie
movies_id_title=[]

# iterate over all movies in dataframe
for idx in movies_with_title.index:
    movies_list.append(movies_with_title['title'][idx])
    movies_id_title.append([movies_with_title['id'][idx],movies_with_title['title'][idx]])
movies_list=set(movies_list)
```

```python
# convert list to dataframe
id_title_df=pd.DataFrame(movies_id_title,columns=['id','title'])

# remove duplicate row
id_title_df=id_title_df.drop_duplicates()
```

```python
# mergin title to new data frame
new=new.merge(id_title_df,on='id')
```

```python
# vectorization for content based fultering (bag of word)
from sklearn.feature_extraction.text import CountVectorizer
cv_bag_of_words = CountVectorizer(max_features=100000,stop_words='english')

# convert whole string into vector means each movies now repersent as a vector
vector_bag_of_words = cv_bag_of_words.fit_transform(new['tags']).toarray()
```

```python
# vectorization for content based fultering (TF-IDF)
from sklearn.feature_extraction.text import TfidfVectorizer
cv_tf_idf = TfidfVectorizer(max_features=100000,stop_words='english')

# convert whole string into vector menas each movies now repersent as a vector
vector_tf_idf=cv_tf_idf.fit_transform(new['overview']).toarray()
```

```python
# for measure angle between vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# generate similarity score
similarity = cosine_similarity(vector_tf_idf)
similarity
```

```python
import pickle
# create a file in dictonary
pickle.dump(similarity,open('similarity.pkl','wb'))
```

```python
# list(enumerate(similarity[index]))  -> this is for making pair of index and cosine angle between two vctor to because after sorting order is not maintained
def recommend_next_10_Movies(movie):
    index=0
    for idx in new.index: 
        if new['title'][idx]==movie:
            index=idx
            break
    # sort all vectors into decreasing order of similarity score of each movie
    nearest_vectors = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    # list of recommed movies 
    List_of_nearest_n_movies_id=[]
    for i in nearest_vectors:
        List_of_nearest_n_movies_id.append(new['title'][i[0]])
    return List_of_nearest_n_movies_id
recommend_next_10_Movies('Iron Man')
```
