```python
# Modules
import numpy as np
import pandas as pd
import ast
import nltk
import pickle 

#importing preprocessed file
movies=pickle.load(open('./data/movies_meta_data.pkl','rb'))

movies
```

```python
# removing all row which having any null value
movies.isnull().sum() 

movies.dropna(inplace=True)
```

```python
# filter out id , user , cast 
movies_new=movies[['id','cast','vote_average']]
movies_new
```

```python
# for storing movie id and cast in that movie id 
id_cast_rating=[]

# for storing movie cast member name
cast_names_list=[]

#iterate over whole dataframe  
for idx in movies_new.index:
    for person in movies_new['crew'][idx]:
        cast_names_list.append(person)
        id_cast_rating.append([movies_new['id'][idx],person,movies_new['vote_average'][idx]])
cast_names_list=set(cast_names_list)
cast_names_list
```

```python
# convert list of id and cast member name into data frame
movies_crew_df=pd.DataFrame(id_cast_rating,columns=['id','person','avg_rating'])
movies_crew_df
```

```python
# remove duplicates from dataframe
movies_crew_df=movies_crew_df.drop_duplicates()
```

```python
# generate a matrix between movie id and cast member with value as average rating of that movie
movies_crew_mat=pd.pivot(movies_crew_df,index='id',columns='person',values='avg_rating').fillna(0)
movies_crew_mat
```

```python
# create folder in current directory
pickle.dump(movies_crew_mat,open('cast_rating.pkl','wb'))
pickle.dump(cast_names_list,open('cast_names.pkl','wb'))
```

```python
# importing similarity matrix and list of movies list for similarity matrix
tfidf_similarity_mat=movies=pickle.load(open('./data/similarity3.pkl','rb'))
movies_for_similarity=pickle.load(open('./data/movies.pkl','rb'))
tfidf_similarity_mat
```

```python
# this fucntion return list of recommendation on the basis of content similarity and it accepts movie name
def recommend_movies_by_movie_name_for_content_based(movie):
    index=0
    for idx in movies_for_similarity.index: 
        if movies_for_similarity['title'][idx]==movie:
            index=idx
            break
    nearest_vectors = sorted(list(enumerate(tfidf_similarity_mat[index])),reverse=True,key = lambda x: x[1])
    list_of_movies=[]
    for i in nearest_vectors:
        list_of_movies.append(movies_for_similarity['id'][i[0]] )
    return list_of_movies
```

```python
# this funciton for recommendation of movie on the basis of person (cast member)
def recommend_movies_by_casting(person_name):
    #  list of movie in which person is appear 
    list_of_highest_rating_movies=[]
    first_movie_id=''
    for idx in (movies_crew_mat[person_name].sort_values(ascending=False)).index:

        # if value is 0 means that person is not appear in that movie
        if movies_crew_mat[person_name][idx]==0.0:
            # if person is not appear in that movie then we further recommend movies on according to the content of first movies which is going to be recommend to that user first
            for i in movies_for_similarity.index:
                if(movies_for_similarity['id'][i]==first_movie_id):
                    # merging both list 
                    list_of_highest_rating_movies=list_of_highest_rating_movies+recommend_movies_by_movie_name_for_content_based(movies_for_similarity['title'][i])[1:]
                    break
            break
        else:
            if len(list_of_highest_rating_movies)==0:
                first_movie_id= idx
            list_of_highest_rating_movies.append(idx)
    return list_of_highest_rating_movies

list_of_movie_id =recommend_movies_by_casting('Jennifer Ehle')
```
